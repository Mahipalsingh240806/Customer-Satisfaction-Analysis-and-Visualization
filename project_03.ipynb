```python
## Encoding

Gender_dict = {'Male' : 0,
                'Female' : 1 }
```

```python
df['Gender'] = df['Gender'].map(Gender_dict)
```

```python
df['Gender']
```

```python
Customer_Type_dict = {'Loyal Customer' : 0,
'disloyal Customer' : 1}
```

```python
df['Customer Type'] = df['Customer Type'].map(Customer_Type_dict)
```

```python
df['Customer Type']
```

```python
Type_of_Travel_dict = {'Personal Travel' : 0,
                        'Business travel' : 1}
```

```python
df['Type of Travel'] = df['Type of Travel'].map(Type_of_Travel_dict)
```

```python
df['Type of Travel']
```

```python
Class_dict = {'Eco Plus' : 0,
'Business' : 1,
'Eco' : 2}
```

```python
df['Class'] = df['Class'].map(Class_dict)
```

```python
df['Class']
```

```python
satisfaction_dict = {'neutral or dissatisfied' : 0,
                    'satisfied' : 1}
```

```python
df['satisfaction'] = df['satisfaction'].map(satisfaction_dict)
```

```python
df['satisfaction']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='satisfaction', data=df)
plt.title("Customer Satisfaction Distribution")
plt.show()
```

```python
sns.countplot(x='Gender', hue='satisfaction', data=df)
plt.title("Gender vs Satisfaction")
plt.show()
```

```python
sns.countplot(x='Customer Type', hue='satisfaction', data=df)
plt.title("Customer Type vs Satisfaction")
plt.show()
```

```python
sns.countplot(x='Type of Travel', hue='satisfaction', data=df)
plt.title("Travel Type vs Satisfaction")
plt.show()
```

```python
sns.countplot(x='Class', data=df)
plt.title("Travel Class Distribution")
plt.show()
```

```python
sns.countplot(x='Class', hue='satisfaction', data=df)
plt.title("Class vs Satisfaction")
plt.show()
```

```python
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['Flight Distance'], bins=20)
plt.title("Flight Distance Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(
    x='satisfaction',
    y='Age',
    data=df
)

plt.show()
```
